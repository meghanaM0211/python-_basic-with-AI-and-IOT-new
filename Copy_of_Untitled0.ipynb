```python
#step-3: Handling the missing values.
#scenario-1: missing values in numerical value
#scenario-2: missing values in non-numerical values
#scenario-3: having missing values in use data save

import numpy as np
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])
```

```python
#step-4:handling catogorial data
#which converts the text data into numeric data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#Alexa-0,0,1

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X=ct.fit_transform(X)
```
